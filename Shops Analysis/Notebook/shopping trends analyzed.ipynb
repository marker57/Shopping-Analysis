## Gender of the mall

```python
labels = ['FEMALE', 'MALE']
size = shops['Gender'].value_counts()

fig, ax = plt.subplots()

plt.pie(size, explode=[0,0.1] ,labels=labels,autopct='%1.2f%%' ,shadow=True,)
plt.title('Gender', fontsize=35,color='red')
plt.legend()
plt.show()
# the are more female in the shops
```

## Category of item purchased

```python
labels=['Clothing','Footwear','Accessories','Outerwear']
clothes=shops['Category'].value_counts()
fig,ax=plt.subplots()
plt.pie(clothes,explode=[0,0,0,0],labels=labels,autopct='1%.2f%%',shadow=True)
plt.title('clothe brand', fontsize=40)
# plt.axis('equal')
plt.xlabel('Gender',color='r',labelpad=20,weight='bold')
plt.legend()
plt.show()
# people wear more of clothing
```

## item purchased by customer

```python
shops['Item_Purchased'].value_counts().sort_values(ascending=False).plot(kind='barh',color=sns.color_palette("Paired"),figsize=(14,8))
plt.title('Item Purchased')
plt.savefig("Images/Item Purchased.png")
plt.xlabel('no of occurences',color='red')
plt.ylabel('clothes',color='red')
plt.show()
```
